```python
import json
class ContactManagement:
    def __init__(self,path='-'):
        self.contact_list=[] #تعریف یه لیست ابتدایی خالی

        if path!='-':
            print('loading previous contacts.....')
            with open(path,'r') as f:
                data=f.read()
                self.contact_list=json.loads(data)
            print('LOADED....')    

    def add(self,name,number):  #یه تابع برای اضافه کردن مخاطب
        self.contact_list.append({
            'name':name,'number':number
        })

    def remove(self,name):
        for items in self.contact_list:   #حلقه برای پیدا کردن اسم در لیست وجود
            if items['name'].lower()==name.lower():  #اگه توی این لیست بود حذف میشه
                self.contact_list.remove(items)
                return f'{name} removed.'
        return f'name {name} not found.'
        
    def search(self,name):  #یه تابع برای جست و جو کردن مخاطب
        result=[]
        for item in self.contact_list:
            if name.lower() in item['name'].lower():
                result.append(item)
                
        if len(result)>0:
            return result
        else:
            return'Not found.'  
        
    def backup(self):  #تابع پشتیبانی
        with open('./contact_list.json','w') as f:   #ایجاد یه فایل json
            f.write(json.dumps(self.contact_list))
            
    def print(self):  #تابع نمایش لیست مخاطبین
        print(f'Your contact lis is: {self.contact_list}')
```

```python
my_contacts=ContactManagement(path='./contact_list.json')
my_contacts.print()
```

```python
my_contacts.add('sara','7878')
my_contacts.print()
```
